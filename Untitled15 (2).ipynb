```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, RMSprop, Adam, Adagrad, Adadelta
from tensorflow.keras.callbacks import EarlyStopping

# Load and preprocess data
data = pd.read_csv('/content/kidney_disease.csv')
data.dropna(inplace=True)

encoder = LabelEncoder()
for col in data.select_dtypes(include=['object']).columns:
    data[col] = encoder.fit_transform(data[col])

X = data.drop('classification', axis=1)
y = data['classification']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Define function to build model
def build_model():
    model = Sequential([
        Dense(32, input_dim=X_train.shape[1], activation='relu'),
        Dense(16, activation='relu'),
        Dense(1, activation='sigmoid')  # Changed to binary classification output
    ])
    return model

# Optimizers to try
optimizers = {
    'SGD': SGD(learning_rate=0.001),
    'SGD+Momentum': SGD(learning_rate=0.001, momentum=0.09),
    'RMSprop': RMSprop(learning_rate=0.0001),
    'Adam': Adam(learning_rate=0.0001),
    'Adagrad': Adagrad(learning_rate=0.001),
    'Adadelta': Adadelta(learning_rate=0.001)
}

results = []
histories = {}

# Early stopping
early_stop = EarlyStopping(patience=10, restore_best_weights=True)

# Train using each optimizer
for name, optimizer in optimizers.items():
    print(f"\nTraining with {name} optimizer...")
    model = build_model()
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

    history = model.fit(X_train, y_train,
                        validation_data=(X_test, y_test),
                        epochs=50,
                        batch_size=8,
                        callbacks=[early_stop],
                        verbose=0)

    train_loss, train_acc = model.evaluate(X_train, y_train, verbose=0)
    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)

    results.append({
        "Optimizer": name,
        "Train Loss": round(train_loss, 4),
        "Train Acc": round(train_acc * 100, 4),
        "Test Loss": round(test_loss, 4),
        "Test Acc": round(test_acc * 100, 2)
    })

    histories[name] = history

# Display Results
df_results = pd.DataFrame(results)
print("\n=== Optimizer Comparison ===")
print(df_results)

# Plotting Loss and Accuracy Curves
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
for name, history in histories.items():
    plt.plot(history.history['val_loss'], label=name)
plt.title("Validation Loss Comparison")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

plt.subplot(1, 2, 2)
for name, history in histories.items():
    plt.plot(history.history['val_accuracy'], label=name)
plt.title("Validation Accuracy Comparison")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Encode categorical columns
import numpy as np

encoder = LabelEncoder()
for col in data.select_dtypes(include=['object']).columns:
    data[col] = encoder.fit_transform(data[col])

# Feature and label separation
X = data.drop('classification', axis=1)
y = data['classification']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standard scaling
scaler = StandardScaler()
# X_train_seq = scaler.fit_transform(X_train)
# X_test_seq = scaler.transform(X_test)

# Before training
y_train = y_train.values if hasattr(y_train, 'values') else y_train
y_test = y_test.values if hasattr(y_test, 'values') else y_test

# Ensure shapes are (n_samples, 1)
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)


X_train = X_train.values if hasattr(X_train, 'values') else X_train
X_test = X_test.values if hasattr(X_test, 'values') else X_test


def sigmoid(x):
    x = np.clip(x, -500, 500)  # Prevent overflow
    return 1 / (1 + np.exp(-x))


# Binary Cross-Entropy Loss
def compute_loss(y_true, y_pred):
    return -np.mean(y_true * np.log(y_pred + 1e-8) + (1 - y_true) * np.log(1 - y_pred + 1e-8))

# Forward Pass
def forward(X, W1, b1, W2, b2):
    hidden_input = np.dot(X, W1) + b1
    hidden_output = sigmoid(hidden_input)
    final_input = np.dot(hidden_output, W2) + b2
    final_output = sigmoid(final_input)
    return final_output, hidden_output

# Initialize Neural Network Parameters
input_neurons = X_train.shape[1]
hidden_neurons = 5
output_neurons = 1

np.random.seed(42)
W1 = np.random.randn(input_neurons, hidden_neurons)
b1 = np.zeros((1, hidden_neurons))
W2 = np.random.randn(hidden_neurons, output_neurons)
b2 = np.zeros((1, output_neurons))

# Hill Climbing Parameters
epochs = 5000
losses = []
step_size = 0.01

# Save best weights
best_W1, best_b1 = W1.copy(), b1.copy()
best_W2, best_b2 = W2.copy(), b2.copy()
best_output, _ = forward(X_train, W1, b1, W2, b2)
best_loss = compute_loss(y_train, best_output)

for epoch in range(epochs):
    # Random perturbations
    delta_W1 = np.random.randn(*W1.shape) * step_size
    delta_b1 = np.random.randn(*b1.shape) * step_size
    delta_W2 = np.random.randn(*W2.shape) * step_size
    delta_b2 = np.random.randn(*b2.shape) * step_size

    # New candidate solution
    W1_new = W1 + delta_W1
    b1_new = b1 + delta_b1
    W2_new = W2 + delta_W2
    b2_new = b2 + delta_b2

    # Forward pass with new weights
    output, _ = forward(X_train, W1_new, b1_new, W2_new, b2_new)
    new_loss = compute_loss(y_train, output)

    # Accept only if better
    if new_loss < best_loss:
        W1, b1 = W1_new, b1_new
        W2, b2 = W2_new, b2_new
        best_loss = new_loss

    losses.append(best_loss)

    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Loss: {abs(best_loss):.4f}")


# Evaluation
final_output_test, _ = forward(X_test, W1, b1, W2, b2)
predictions = (final_output_test > 0.5).astype(int)
accuracy = np.mean(predictions == y_test) * 100

print(f"\nTest Accuracy: {accuracy:.2f}%")

# Plot Loss Curve
plt.plot(losses)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Curve with Hill Climbing')
plt.show()
```

```python
import numpy as np

# Assume 'data' is a cleaned pandas DataFrame already loaded before this

# Manual encoding
data['classification'] = data['classification'].replace({'ckd': 1, 'notckd': 0})
for col in data.columns:
    if data[col].dtype == 'object' and col != 'classification':
        data[col] = pd.factorize(data[col])[0]

# Prepare X and y
X = data.drop('classification', axis=1).values
y = data['classification'].values.reshape(-1, 1)

# Normalize features
mean = X.mean(axis=0)
std = X.std(axis=0)
X = (X - mean) / std

# Split
split = int(0.7 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Activation functions
def sigmoid(x):
    x = np.clip(x, -500, 500)
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

# Loss function
def compute_loss(y_true, y_pred):
    y_pred = np.clip(y_pred, 1e-7, 1 - 1e-7)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# Forward pass
def forward(X, W1, b1, W2, b2, W3, b3):
    h1 = relu(np.dot(X, W1) + b1)
    h2 = relu(np.dot(h1, W2) + b2)
    out = sigmoid(np.dot(h2, W3) + b3)
    return out, h1, h2

# Init weights
input_neurons = X.shape[1]
hidden1 = 25
hidden2 = 32
output_neurons = 1

np.random.seed(42)
W1 = np.random.randn(input_neurons, hidden1) * np.sqrt(2. / input_neurons)
b1 = np.zeros((1, hidden1))
W2 = np.random.randn(hidden1, hidden2) * np.sqrt(2. / hidden1)
b2 = np.zeros((1, hidden2))
W3 = np.random.randn(hidden2, output_neurons) * np.sqrt(2. / hidden2)
b3 = np.zeros((1, output_neurons))

# Hill Climbing
epochs = 500
step_size = 0.01
losses = []

best_W1, best_b1 = W1.copy(), b1.copy()
best_W2, best_b2 = W2.copy(), b2.copy()
best_W3, best_b3 = W3.copy(), b3.copy()

best_output, _, _ = forward(X_train, W1, b1, W2, b2, W3, b3)
best_loss = compute_loss(y_train, best_output)

for epoch in range(epochs):
    delta_W1 = np.random.randn(*W1.shape) * step_size
    delta_b1 = np.random.randn(*b1.shape) * step_size
    delta_W2 = np.random.randn(*W2.shape) * step_size
    delta_b2 = np.random.randn(*b2.shape) * step_size
    delta_W3 = np.random.randn(*W3.shape) * step_size
    delta_b3 = np.random.randn(*b3.shape) * step_size

    W1_new = W1 + delta_W1
    b1_new = b1 + delta_b1
    W2_new = W2 + delta_W2
    b2_new = b2 + delta_b2
    W3_new = W3 + delta_W3
    b3_new = b3 + delta_b3

    output, _, _ = forward(X_train, W1_new, b1_new, W2_new, b2_new, W3_new, b3_new)
    new_loss = compute_loss(y_train, output)

    if new_loss < best_loss:
        W1, b1 = W1_new, b1_new
        W2, b2 = W2_new, b2_new
        W3, b3 = W3_new, b3_new
        best_loss = new_loss

    losses.append(best_loss)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {best_loss:.6f}")

# Final test eval
final_output, _, _ = forward(X_test, W1, b1, W2, b2, W3, b3)
preds = (final_output > 0.5).astype(int)
accuracy = np.mean(preds == y_test) * 100

print(f"\nTest Accuracy: {accuracy:.2f}%")

# Optional: Print confusion matrix manually
TP = np.sum((preds == 1) & (y_test == 1))
TN = np.sum((preds == 0) & (y_test == 0))
FP = np.sum((preds == 1) & (y_test == 0))
FN = np.sum((preds == 0) & (y_test == 1))

print(f"\nConfusion Matrix:")
print(f"TP: {TP}, FP: {FP}")
print(f"FN: {FN}, TN: {TN}")
```
